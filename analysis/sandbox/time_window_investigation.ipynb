# Look at the Different Time Windows

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib widget
```

```python
# Window 1
window1df = pd.read_csv('../../data/play1_window1_df.csv')
window2df = pd.read_csv('../../data/play1_window2_df.csv')
window3df = pd.read_csv('../../data/play1_window3_df.csv')
window4df = pd.read_csv('../../data/play1_window4_df.csv')
window5df = pd.read_csv('../../data/play1_window5_df.csv')
```

```python
window1df['swiftnum']
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))
ax1.scatter(window1df['cross ice coord'], window1df['along ice drift speed'])
ax1.scatter(window2df['cross ice coord'], window2df['along ice drift speed'])
ax1.scatter(window3df['cross ice coord'], window3df['along ice drift speed'])
ax1.scatter(window4df['cross ice coord'], window4df['along ice drift speed'])
ax1.scatter(window5df['cross ice coord'], window5df['along ice drift speed'])
ax1.set_xlabel('Cross Ice Location [m]')
ax1.set_ylabel('Along Ice Drift Speed [m/s]')

ax2.scatter(window1df['cross ice coord'], window1df['cross ice drift speed'])
ax2.scatter(window2df['cross ice coord'], window2df['cross ice drift speed'])
ax2.scatter(window3df['cross ice coord'], window3df['cross ice drift speed'])
ax2.scatter(window4df['cross ice coord'], window4df['cross ice drift speed'])
ax2.scatter(window5df['cross ice coord'], window5df['cross ice drift speed'])
ax2.set_xlabel('Cross Ice Location [m]')
ax2.set_ylabel('Cross Ice Drift Speed [m/s]')

plt.tight_layout()
```

Lets try normalizing the along ice drift speed by the average open ocean along ice drift speed. Maybe subtract out the drift due to wind? Is that possible to do?

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10,5))
ax1.scatter(window1df['cross ice coord'], window1df['along ice drift speed'], c=window1df['swiftnum'], cmap='Accent')
ax1.scatter(window2df['cross ice coord'], window2df['along ice drift speed'], c=window2df['swiftnum'], cmap='Accent')
ax1.scatter(window3df['cross ice coord'], window3df['along ice drift speed'], c=window3df['swiftnum'], cmap='Accent')
ax1.scatter(window4df['cross ice coord'], window4df['along ice drift speed'], c=window4df['swiftnum'], cmap='Accent')
ax1.scatter(window5df['cross ice coord'], window5df['along ice drift speed'], c=window5df['swiftnum'], cmap='Accent')
ax1.set_xlabel('Cross Ice Location [m]')
ax1.set_ylabel('Along Ice Drift Speed [m/s]')

ax2.scatter(window1df['cross ice coord'], window1df['cross ice drift speed'], c=window1df['swiftnum'], cmap='Accent')
ax2.scatter(window2df['cross ice coord'], window2df['cross ice drift speed'], c=window2df['swiftnum'], cmap='Accent')
ax2.scatter(window3df['cross ice coord'], window3df['cross ice drift speed'], c=window3df['swiftnum'], cmap='Accent')
ax2.scatter(window4df['cross ice coord'], window4df['cross ice drift speed'], c=window4df['swiftnum'], cmap='Accent')
ax2.scatter(window5df['cross ice coord'], window5df['cross ice drift speed'], c=window5df['swiftnum'], cmap='Accent')
ax2.set_xlabel('Cross Ice Location [m]')
ax2.set_ylabel('Cross Ice Drift Speed [m/s]')

plt.tight_layout()
```
