# SASSIE SWIFT exploration 

```python
import numpy as np
import matplotlib.pyplot as plt
import netCDF4 as nc

%matplotlib widget
```

```python
# Load play 4 data
swift12_play4_dataset = nc.Dataset('../swift-data/SASSIE_Fall_2022_SWIFT12_play4.nc')
swift12_play4_dataset
```

```python
# Plot the SWIFT trajectory
lat = swift12_play4_dataset['latitude'][:]
lon = swift12_play4_dataset['longitude'][:]

fig, ax = plt.subplots()
ax.scatter(lat, lon)
```

```python
time = swift12_play4_dataset['time'][:]
fig, (ax1, ax2) = plt.subplots(nrows=2)
ax1.plot(time, lat)
ax1.set_ylabel('lat')
ax2.plot(time, lon)
ax2.set_ylabel('lon')
```

```python
hs = swift12_play4_dataset['surface_wave_height'][:]
fig, ax = plt.subplots()
ax.plot(time, hs)
```

```python
fig, ax = plt.subplots()
im = ax.scatter(lon, lat, c=hs, cmap='inferno')
cbar = fig.colorbar(im)
cbar.set_label('Significant wave height [m]')
```
