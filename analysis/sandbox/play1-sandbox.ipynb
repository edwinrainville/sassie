```python
import cmocean
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib widget
```

```python
# Load the Play 1 Dataframe
df = pd.read_csv('../../data/play1_df.csv')
df.head()
```

```python
# Wave Height Variations with Ice Concentration
fig, ax = plt.subplots()
im = ax.scatter(df['cross ice coord'], df['hs'], c=df['ice concentration'], cmap=cmocean.cm.ice, edgecolors='k')
cbar = fig.colorbar(im)
cbar.set_label('Ice Conentration [%]')
ax.set_xlabel('Cross Ice Distance [m]')
ax.set_ylabel('Significant Wave Height [m]')
```

Well nothing too surprising here. More attenuation with higher ice concentration. The ice aligns with the ice edge that was picked out as expected. Steeper attenuation with higher ice concentration.
